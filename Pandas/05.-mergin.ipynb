# **Combinaci√≤n de DataFrames**

## **Concatenacion**

```python
data_one = { 'A': ['A0', 'A1', 'A2', 'A3'], 'B': ['B0', 'B1', 'B2', 'B3'] }
```

```python
data_two = { 'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3'] }
```

```python
one = pd.DataFrame(data_one)
```

```python
two = pd.DataFrame(data_two)
```

```python
one
```

```python
two
```

```python
# concatenacion por columnas
pd.concat([one, two], axis=1)
```

```python
# concatenacion por filas
pd.concat([one, two], axis=0)
```

```python
two.columns = one.columns
```

```python
two
```

```python
pd.concat([one, two], axis=0)
```

# **Merging**

<ul>
    <li>inner merge</li>
    <li>outer merge</li>
    <li>Left o Right merge</li>
</ul>

```python
registration = pd.DataFrame( {'reg_id': [1, 2, 3, 4], 'name': ['Andrew', 'Bob','Claire', 'David']} )
login = pd.DataFrame( {'log_id': [1, 2, 3, 4], 'name': ['Xavier', 'Andrew', 'Yolanda', 'Bob']} )
```

```python
registration
```

```python
login
```

<ul>
    <li><h2>Inner Merge</h2></li>
</ul>

```python
pd.merge(registration, login, how='inner', on='name')
```

```python
pd.merge(login, registration, how='inner', on='name')
```

<ul>
    <li><h2>Left y Right Merge</h2></li>
</ul>

<p> <em>En el caso del Left Merge y Right Merge el orden de las tablas importa </em> </p>

```python
# left merge
pd.merge(left=registration, right=login, how='left', on='name')
```

```python
pd.merge(left=login, right=registration, how='left', on='name')
```

```python
pd.merge(left=registration, right=login, how='right', on='name')
```

<ul>
    <li><h2>Outer Merge</h2></li>
</ul>

```python
pd.merge(registration, login, how='outer', on='name')
```

```python
registration = registration.set_index('name')
```

```python
registration
```

```python
login
```

```python
pd.merge(registration, login, left_index= True, right_on= 'name', how= 'inner')
```

```python
registration = registration.reset_index()
```

```python
registration
```

```python
registration.columns = ['reg_name','reg_id']
```

```python
registration
```

```python
login
```

```python
pd.merge(registration, login, how='inner', left_on='reg_name', right_on= 'name')
```

```python
registration.columns = ['name', 'id']
```

```python
registration
```

```python
login
```

```python
pd.merge(registration, login, how='inner', on='name', suffixes=['_reg', '_log'])
```
