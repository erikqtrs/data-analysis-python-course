<h1>SQL Databases</h1>

Pandas puede leer y escribir en algunos motores de Base de Datos

mediante el uso de un driver y de la libreria **sqlalchemt**

<h2>Pasos para manejar SQL a traves de Pandas</h2>
<ul>
    <li><h3><strong>Paso 1:</strong></h3></li>
    <p> Averigua a que motor SQL estas conectado, como:</p>
    <ul>
        <li>PostgreSQL</li>
        <li>MySQL</li>
        <li>MS SQL Server</li>
    </ul>
</ul>

<ul>
    <li><h3><strong>Paso 2:</strong></h3></li>
    <p> Instala el driver o libreria o conector de Python correspondiente a tu BBDD </p>
    <ul>
        <li>PostgreSQL - psycopg2</li>
        <li>MySQL - pymysql o mysql-connector-python </li>
        <li>MS SQL Server - pyodbc</li>
    </ul>
</ul>

<ul>
    <li><h3><strong>Paso 3:</strong></h3></li>
    <ul>
        <li>Usar sqlalchemy para conectarse con la base de datos SQL</li>
    </ul>
</ul>

<ul>
    <li><h3><strong>Paso 4:</strong></h3></li>
    <ul>
        <li>Usa la coneccion de sqlalchemy driver con el metodo de pandas read_sql() </li>
        <li>Pandas puede leer toda una tabla como DataFrame o pasar una query SQL </li>
    </ul>
</ul>

```python
from sqlalchemy import create_engine
import numpy as np
```

```python
temp_db = create_engine('sqlite:///:memory:')
```

```python
df = pd.DataFrame(data=np.random.randint(low=0, high=100, size=(4, 4)), columns=['a', 'b', 'c', 'd'])
```

```python
df
```

```python
df.to_sql(name='new_table', con=temp_db)
```

```python
new_df = pd.read_sql(sql='new_table', con=temp_db)
```

```python
new_df
```

```python
pd.read_sql_query(sql= 'SELECT a,c FROM new_table', con=temp_db)
```
