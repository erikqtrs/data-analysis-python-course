# Pandas

## Series

- El indice nombrado lo diferencia de un ndarray simple de Numpy

### Trabajando con Series

```python
# import numpy as np
```

```python
import pandas as pd
```

```python
# help( pd.Series )
```

```python
myindex = ['Ecuador', 'Colombia', 'Perú']
```

```python
mydata = [2140, 4175, 5824]
```

```python
# array normal, indice numerico
myserie = pd.Series(data=mydata)
myserie
```

```python
# Serie con indexado etiquetado
myserie = pd.Series( data=mydata, index=myindex )
myserie
```

### Indexacion de una Serie como diccionario de Python

```python
myserie[1]
```

```python
myserie['Perú']
```

```python
# añadir elementos
myserie['Venezuela'] = 2147
```

```python
myserie
```

```python
# eliminar elementos
del myserie['Perú']
myserie
```

```python
ages = {'Sam': 18, 'Veronica': 14, 'Frank': 8}
myserie = pd.Series(ages)
```

```python
myserie
```

```python
# Datos imaginarios sobre ventas del 1er y 2do cuarto de una empresa XYZ
q1 = {'Japon': 80, 'China': 450, 'India': 200, 'Estados Unidos': 250}
q2 = {'Brasil': 100, 'China': 500, 'India': 240, 'Estados Unidos': 120}
```

```python
sales_q1 = pd.Series(q1)
```

```python
sales_q2 = pd.Series(q2)
```

```python
sales_q1
```

```python
sales_q2
```

```python
# comprovar si una clave existe
'japan' in sales_q1
```

#### **Nota:** Las claves en las Series de pandas son case-sensitive

```python
# obtener claves
sales_q1.keys()
```

```python
# obtener valores de la serie
sales_q1.values
```

```python
# suma entre series
sales_q1 + sales_q2
```

```python
sales_q1.add(sales_q2, fill_value=0)
```
