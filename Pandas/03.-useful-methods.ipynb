## Aplicar una Función personalizada a un DataFrame

- **Método apply()**

```python
def last_four(num):
    """ Esta función extrae los ultimos caracteres
    de un string """
    return str(num)[-4:]
```

```python
last_four(142354)
```

```python
df['last_four'] = df['CC Number'].apply(last_four)
```

```python
df['total_bill'].mean()
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 20:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```

## apply() que usa varias columnas

```python
df.head()
```

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
```

```python
quality(16.9, 1.02)
```

```python
df['Quality'] = df[['total_bill', 'tip']].apply( lambda df: quality(df['total_bill'], df['tip']), axis = 1 )
```

## Vectorizar una serie

- **np.vectorize(function)**

El propòsito de np.vectorize() es tranformar funciones que no pueden ser 

tratadas o ejecutadas con numpy, en funciones que pueden ser ejecutadas como

un array de numpy.

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```

```python
df.head()
```

## **Comparaciòn de ejecucion en tiempo**

Para poder comparar el tiempo de ejecucion se utilizara el modulo **timeit**

el cual necita de dos cosas:

- setup: variable string con los modulos y funcion a analizar
- stmt: statement o variables con el codigo a analizar

```python
import timeit
```

```python
# setup se corre una sola vez
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('./assets/tips.csv')
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
'''
```

```python
# statements se corre una sola vez
stmt_one = '''
df['Tip Quality'] = df[['total_bill', 'tip']].apply( lambda df: quality(df['total_bill'], df['tip']), axis = 1 )
'''

stmt_two = '''
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
```

```python
timeit.timeit(setup=setup, stmt=stmt_one, number=1000)
```

```python
timeit.timeit(setup=setup, stmt=stmt_two, number=1000)
```
