## Conditional Filtering

```python
df = pd.read_csv('./assets/tips.csv')
```

```python
df
```

```python
df['total_bill'] > 40
```

```python
# ver todas las filas
#pd.options.display.max_rows = None
```

```python
df['total_bill'] > 40
```

- ## Ver los valores que cumples la condicion

```python
bool_series = df['total_bill'] > 40
```

```python
df[bool_series]
```

```python
# forma corta de mostrar valores que cumplen condicion
df[ df['total_bill'] > 40 ]
```

```python
df[df['sex'] == 'Male']
```

- ## Filtrado condicional con operadores booleanos

AND &      -------------        Las condiciones deben ser verdaderas

OR |       --------------       Una de las condiciones debe ser verdadera

```python
df[ (df['total_bill'] > 40) & (df['sex'] == 'Female') ]
```

```python
df[ (df['smoker'] == 'Yes') | (df['sex'] == 'Male') ]
```

```python
# ver facturas de sabado, domingo, viernes
df[ (df['day'] == 'Sun') | (df['day'] == 'Sat') | (df['day'] == 'Fri')]
```

- ## Forma corta de hacer un OR

Asi como en SQL tenemos la clausula **IN**, en Pandas podemos usar el metodo

**isin()** para realizar un OR de forma corta.

```python
options = ['Sun', 'Sat', 'Fri']
```

```python
df[df['day'].isin(options)]
```
