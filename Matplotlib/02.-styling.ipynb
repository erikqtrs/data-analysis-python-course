<h1>Estilos en Matplotlib</h1>

Matplotlib cuenta con varias funciones para dar estilos a los graficos, estilos como:
etiquetas, leyendas, colores, etc

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure(figsize=(3, 2), dpi = 150)

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')

ax.plot(x, x**2, label = 'x vs X^2')
ax.legend(loc=(1.2, 0.5))
```

```python
x = np.arange(0, 20, 2)
```

```python
fig = plt.figure(figsize=(3, 2), dpi = 150)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x,x, color='red', lw= 3, ls = '--', label='x V x', marker = 's', ms=12,\
    markerfacecolor='black', markeredgewidth=4, markeredgecolor= 'orange') # RGB HEX Code
ax.plot(x, x+2, color= 'purple', lw = 1.5, ls= '-.', marker= '+', label= 'x V x+2')
ax.legend()
```
