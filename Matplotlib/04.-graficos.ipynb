# Graficos en Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 12, 13)
y = np.random.randint(4, 25, size=13)
```

## Grafico de Linea

Los graficos de linea nos permiten observar cambios en relacion al tiempo, patrones

o tendencias de una variable

```python
fig, ax = plt.subplots(nrows=1,ncols=1)
ax.plot(x, y, color='blue', lw=6)
```

## Grafico de Barras

El grafico de barras es uno de los graficos mas usados en el analisis estadistico

ya que es muy facil de interpretar, este grafico sirve para representar o comparar las

cantidades de una variable en diferente tiempo o comparar varias variables en un mismo tiempo, por

ejemplo, la cantidad de ventas en diferentes a√±os

```python
plt.bar(x, x**2, color='red')
```

## Histograma

Un histograma representa las frecuencias de una distribucion o conjunto de datos,

se utiliza en ciencias sosciales, humanas y economicas.

En si el histograma representa graficamente el numero de veces que aparece un dato.

```python
estatura =np.array([1.80, 1.60, 1.65, 1.75, 1.80, 1.70, 1.75, 1.89, 1.55,
1.78, 1.90, 1.78, 1.65, 1.94, 2.02, 2.10, 1.92, 1.74, 1.84, 1.73, 1.58, 1.98,
1.97, 2.05, 1.74, 1.57, 1.72, 1.83, 1.94])

plt.hist( estatura, 12, color='yellow', edgecolor='black' )
plt.title('Frecuencia de estaturas')
```

## Diagrama de Torta

El diagrama de torta o pastel o pie, es un grafico que nos permite

representar porciones de una variable, o porcentajes

```python
data = pd.DataFrame({
    'nombre': ['Ana', 'Juan', 'Diana', 'Catalina'],
    'manzanas': [20, 10, 25, 30]
})
```

```python
data
```

```python
plt.pie( data.manzanas, labels= data.nombre)
plt.legend(loc=(1.1, 0.5))
plt.show()
```

- ### Agregando porcentajes

```python
plt.pie( data.manzanas, labels= data.nombre, autopct='%0.1f %%')
plt.legend(loc=(1.1, 0.5))
plt.show()
```

- ### Cambiando colores

```python
colores = ['yellow', 'orange', 'lightblue', 'lightgreen']
plt.pie( data.manzanas, labels= data.nombre, autopct='%0.1f %%', colors= colores)
plt.legend(loc=(1.1, 0.5))
plt.show()
```

```python
import matplotlib.colors as mcl
import matplotlib.cm as cm
normdata = mcl.Normalize(min(data.manzanas), max(data.manzanas))
colormap = cm.get_cmap('Reds')
colores = colormap(normdata(data.manzanas))

plt.pie( data.manzanas, labels= data.nombre, autopct='%0.1f %%', colors= colores)
plt.legend(loc=(1.1, 0.5))
plt.show()
```

- ### Separar porciones

```python
desface = [0.1, 0.2, 0.1, 0.4]
colores = ['yellow', 'orange', 'darkred', 'lightgreen']
plt.pie( data.manzanas, labels= data.nombre, autopct='%0.1f %%', colors= colores, explode= desface)
plt.legend(loc=(1.1, 0.5))
plt.show()
```

- ### Cambiando propiedades de texto

```python
colores = ['darkred', 'darkblue', 'darkgreen', 'darkorange']
plt.pie( data.manzanas, labels= data.nombre, autopct='%0.1f %%', 
        colors= colores, textprops={'color': 'white', 'fontsize': 12})
plt.legend(loc=(1.1, 0.5), title='Alumnos')
title = plt.title('Apple Pie Portions')
plt.setp(title, color='white', size=12, weight = 'bold')
plt.show()
```

## Diagrama de dispersion

Un diagrama de dispersion nos permite observar la relacion o combinacion que

existe entre dos variables

```python
x = np.linspace(0, 20, 30)
y = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows= 1, ncols= 1)
fig.set(figwidth = 10, figheight = 5, dpi= 125)
ax.scatter( x, y, marker='x', color='blue' )
```

```python
age_car =  [2, 8, 10, 4, 6, 20, 3, 7, 8, 9, 16]
speed_car = [40, 45, 60, 120, 80, 110, 87, 130, 50, 240, 185]

fig, ax = plt.subplots(nrows= 1, ncols= 1)
fig.set(dpi=150)
ax.scatter(age_car, speed_car, color='blue')
ax.set_xlabel('Age CAr')
ax.set_ylabel('Speed Car')
```
