```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(1, 10)
```

```python
y = 2 * x
```

```python
x
```

```python
y
```

<h2>Functional Programming Method in Matplotlib</h2>

```python
# crear grafico
plt.plot(x, y)                              # Crea el grafico
plt.title('String Line')                    # Titulo del grafico
plt.xlabel('X Axis')                        # Etiqueta del eje X
plt.ylabel('Y Axis')                        # Etiqueta del eje Y
plt.xlim(0, 6)                              # Intervalo del eje X
plt.ylim(0, 15)                             # Intervalo de eje Y
plt.savefig('./assets/myfirstplot.png')     # Guardamos la figura como png
```

<h2>OOP Method in Matplotlib</h2>

<ul type= 'square'>
    <li>La API de POO de Matplotlib utiliza un objeto para una mejor comprension</li>
    <li>Luego añadimos los ejes al objeto Figure y luego graficamos en esos ejes</li>
    <li>Esto permite un control mas robusto sobre todo el grafico</li>
</ul>

```python
# DATA
```

```python
a = np.linspace(0, 10, 11)
```

```python
b = a ** 4
```

```python
a
```

```python
b
```

```python
x = np.arange(0, 10)
```

```python
y = 2 * x
```

```python
x
```

```python
y
```

```python
fig = plt.figure()                      # Crear el lienzo
# LARGE AXES
axes1 = fig.add_axes([0, 0, 1, 1])       # Agregar los ejes
axes1.plot(a, b)                         # Realizando la figura o grafico
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# SMALL AXES
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a, b)
axes2.set_xlim(2, 3)
axes2.set_ylim(0, 500)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

plt.show()
```

<h2>Parametros del Objeto Figure</h2>

<ul type='square'>
    <li><strong> num:</stron> numero identificador de la figura</li>
    <li><strong>figsize:</stron> tamaño de la figura (ancho, alto) en pulgadas</li>
    <li><strong> dpi:</stron> resolusion de la figura (dots-per-inch)/li>v
    <li><strong> facecolor:</stron> fondo de la figura color en rgb, rgba, hex, strng</li>
    <li><strong> edgecolor:</stron> color de los bordes de la figura</li>
</ul>

```python
fig = plt.figure(figsize=(2, 2), dpi=200, facecolor='salmon')

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

fig.savefig('./assets/figure-plot.png')
```

<h1>Subplots</h1>

En teoria podemos crear un objeto Figure y manualmente añadir un arreglo de ejex, pero Matplotlib

ya trae consigo una funcion o metodo que hace esto automaticamente por nosotros **plt.subplots()**

```python
fig, axes = plt.subplots( nrows=1, ncols= 2 )
fig.set_dpi(120)
axes[0].plot(x, y)
axes[1].plot(a, b)
```

```python
fig, axes = plt.subplots( nrows=2, ncols= 2 )
fig.set_dpi(150)

axes[0][0].plot(x, y)
axes[1][1].plot(a, b)
```

```python
fig, axes = plt.subplots( nrows= 3, ncols= 1 )
fig.set_dpi(120)

for ax in axes:
    ax.plot(x, y)
```

```python
fig, axes = plt.subplots( nrows= 2, ncols= 2, figsize= (5, 4), dpi= 200 )

axes[0][0].plot(x, y)
axes[0][0].set_xlabel('X LABEL')
axes[0][1].plot(x, y)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)
axes[1][1].set_title('Power 4')
fig.subplots_adjust(wspace=0.8, hspace=0.5)
fig.suptitle('Figure Level', fontsize= 16)

fig.savefig('./assets/subplots.png')
```
