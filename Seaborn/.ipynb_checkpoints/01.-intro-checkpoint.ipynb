# Seaborn

Seaborn es una libreria de visualizacion estadistica de Python, basada en Matplotlib

a diferencia de matplotlib, seaborn nos permite trabajar de manera sencilla

con DataFrame de Pandas, su sintaxis es muy sencilla ya que es one-line

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
#from jupyterthemes import jtplot
#jtplot.style(theme= 'onedork')
```

# Scatter Plots

```python
df = pd.read_csv('data/dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize= (12, 4), dpi= 200)
sns.scatterplot(x= 'salary', y= 'sales', data= df, hue= 'level of education',
                   palette= 'Dark2', alpha= 0.9, s= 200, style= 'level of education' )
plt.savefig('scatterplot.png')
```

# Graficos de Distribucion

Este tipo de graficos despliegan una sola caracteristica continua y ayuda

a visualizar propiedades como desviacion y promedio

### Hay 3 tipos principales de graficos de distribucion:
- Rug Plot
- Histograma
- KDE Plot

```python
df.head()
```

# Rug Plot

```python
sns.rugplot( x= 'salary', data= df, height= 0.5 )
```

# Histogram

```python
sns.displot( data= df, x= 'salary', bins= 20, color= 'blue',
           kde= True, rug= True)
```

# KDE Plot

```python
sns.kdeplot( data= df, x= 'salary' )
```

```python
sample_ages = np.random.randint(0, 100, 200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame( sample_ages, columns= ['age'] )
```

```python
sns.displot( data= sample_ages, x= 'age', bins= 30 )
```

```python
sns.kdeplot( data= sample_ages, x= 'age', bw_adjust= 0.1, color= 'green' )
```

## Categorical Plots

- countplot()
<pre>   Muestra el numero de elementos dentro de una categoria</pre>
- barplot()
- boxplot()
- violinplot()
- swarmplot()
- boxenplot()

```python
df.head()
```

```python
df.division.value_counts()
```

```python
plt.figure( figsize= (10, 5), dpi= 200 )
plt.title('Sales By Division')
sns.countplot( data= df, x= 'division' )
```

```python
df['level of education'].value_counts()
```

```python

plt.figure( figsize= (10, 5), dpi= 200 )
sns.countplot( data= df, x= 'level of education', hue= 'division' ) 
```

```python
# Esto representa el grafico de arriba
df.pivot_table( index= ['level of education', 'division'], aggfunc='count')
```

```python
plt.figure( figsize= (10, 5), dpi= 200 )
sns.barplot( x= 'level of education' , y= 'salary', data= df, estimator=np.mean)
```

```python
df = pd.read_csv( './data/StudentsPerformance.csv' )
```

```python
df
```

```python
plt.figure( figsize= (10, 5), dpi= 200 )
sns.boxplot( x='test preparation course', y= 'math score', data= df )
```

```python
plt.figure( figsize= (10, 5), dpi= 200 )
sns.boxplot( x='parental level of education', y= 'reading score', data= df,
           hue= 'test preparation course')
```

```python
plt.figure( figsize= (10, 8), dpi= 200 )
sns.violinplot( x='reading score', y= 'parental level of education', data= df,
           hue= 'test preparation course', palette= 'inferno', split= True, inner= None)
```

```python
plt.figure( figsize= (8, 4), dpi= 200 )
sns.swarmplot( x= 'math score', y= 'gender', data= df, size= 2 )
```

### Grid Plots

```python
plt.figure( figsize= (10, 5), dpi= 150 )
sns.catplot( x= 'gender', y= 'math score', data= df, kind= 'box',
           col= 'lunch')
```
